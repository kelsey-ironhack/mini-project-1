```python
def mostrartablero(tablero):
    for fila in tablero:
        print('_'*13)#aqui imprimimos las lineas
        for j in fila:
            print('| ',j, ' ', end='', sep='')#Aquí le damos formato al tablero, espacio para cada número y barras verticales
        print('|')                                     #123
    print('_'*13) #Este es el cierre del tablero       #456
                                                       #789    

def jugada(tablero, posicion, jugador): #Esto debe ser una devolución buleana
    posicion -= 1 #Lo ponemos así porque son indices de base cero
    esta_ocupado = tablero[posicion // 3][posicion % 3] == 'X' or tablero[posicion // 3][posicion % 3] == 'O' #Aquí va a comparar si para checar que no se repitan las casillas
    if esta_ocupado:
        return False
    else:
        tablero[posicion // 3][posicion % 3] = jugador
        return True
    
    
def evaluacion(tablero):
    f0 = f1 = f2 = c0 = c1 = c2 = d1 = d2 = True
    for i in range(2):
        f0 = f0 and tablero[0][i] == tablero[0][i + 1] # estamos comparando la primera celda con la segunda y si son iguales devolveran true
        f1 = f2 and tablero[1][i] == tablero[1][i + 1] 
        f2 = f2 and tablero[2][i] == tablero[2][i + 1] 
        c0 = c0 and tablero[i][0] == tablero[i + 1][0]
        c1 = c1 and tablero[i][1] == tablero[i + 1][1]
        c2 = c2 and tablero[i][2] == tablero[i + 1][2]
        d1 = d1 and tablero[i][i] == tablero[i + 1][i + 1]
        d2 = d2 and tablero[2 - i][i] == tablero[2 - i - 1][i + 1]
    return f0 or f1 or f2 or c0 or c1 or c2 or d1 or d2    

        
base = []
fila = []
for i in range(1, 10):
    fila.append(i)
    if i % 3 == 0: # esto implica que primero en 3 primero va agregar el utlimo valor y luego va a evaluar esta condicio
        base.append(fila)
        fila = [] # en caso de que sea 0 resetamos la fila

tablero = base[:]#aqui lo copiamos
jugador = ('X','O')#El jugador será una tupla porqué no se va alterar y sólo podra jugar en posicion (0,1)(X,O) de la tupla
contador = 0 #el contador nos iniciado en 0 nos dice que x será quien inicie las partidas
mensaje = True
while True: #Con este while el tablero se mostrara después de cada jugada
    
    if mensaje:
        mostrartablero(tablero)
        print('Juega', jugador[contador % 2], end=' ')# Aquí hacemos que itere para que cambie entre X y O, colocamos un 'end' para que no haga un enter
        posicion = int(input(' ingrese posición: '))
    else: #Este else es para cuando mensaje=true no se cumpla
        print('posición invalida', end='')
        posicion = int(input('vuelva a ingresar posición: '))
    mensaje = jugada(tablero, posicion, jugador[contador % 2])# Mensaje debe tomar el valor de "jugada "
    if evaluacion(tablero):
        mostrartablero(tablero)
        print('Ganador', jugador[contador % 2])
        break
    if contador == 8:
        print('Empate, nadie gana')
        break 
    if  mensaje: #Aquí indicamos que el jugador es la tupla en la posicion de contador para que este iterando entre los valores 0 y 1
         contador += 1
        
        
```
